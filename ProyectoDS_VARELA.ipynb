## Limpiando missing values

Haciendo un primer análisis vemos que el mayor numero posible de entradas por artributo es de 146660 y que todos cumplen esa condición exeptuando los siguientes:


* lat -	9925 *(missing)*
* lon -	9959 *(missing)*
* bathrooms -	5957 *(missing)*
* surface_total -	20527 *(missing)*
* surface_covered -	21614 *(missing)*

no nos interesa preservar entradas que no sean geolocalizables por lo que vamos a eliminar del set todos los casos que no dispongan de ubicación, en el caso de baños vamos a reemplazar los valores fantantes por 0 porque no todas las propiedades son casas, por lo que la falta de ese dato no afecta la consistencia del set, en el caso de surface_total y surface_covered intentaremos reemplazar la información faltante con un valor que sea función de la cantidad de habitaciones (rooms) listadas en la entrada particular, mientras que en el atributo surface_total, los valores faltante se calcularan adicionando un 0.07 de superficie al valor de surface_covered.

```python

# Quitando entradas con valores faltantes de latitud y longitud
df.dropna(subset=['lat', 'lon'], inplace=True)
df.isnull().sum()

```

```python
#Reemplazamos en el atributo bathrooms las entradas faltantes con valor 0
df["bathrooms"].fillna(0, inplace=True)
df.isnull().sum()


```

```python
#Hay valores en superficie cubierta que son mayores a los de superficie total, vamos a encontrar esos casos y quitarlos de la base

count = 0

for index, row in df.iterrows():
  if row['surface_covered'] > row['surface_total']:
    print(index, row['surface_covered'],row['surface_total'])
    df.drop(index, inplace=True)
    count = count +1

print('Cantidad de Casos:', count)

```

```python
#presentando valores nulos con atencion en Surface Total
df.isnull().sum()
```

```python
#Reemplazamos en el atributo surface_covered las entradas faltantes un valor que sea una función de la cantidad de habitaciones mas un 7% de superficie

sel = df['surface_total'].isnull() & df['surface_covered'].notnull()
df.loc[sel, 'surface_total'] = df.loc[sel, 'surface_covered'] * 1.07

df.isnull().sum()

#tendriamos que recuperar 486 casos 19122 --> 18636
```

```python
# Dropeamos el resto de las entradas que no tienen definida surface_total ya que lo mas probable es que sean errores en la carga

df.loc[df['surface_total'].isnull()].head(10)
df.dropna(subset=['surface_total'], inplace=True)
df.isnull().sum()
```

```python
# Por ultimo reemplazamos todos los valores nulos de surface covered por 0

df["surface_covered"].fillna(0, inplace=True)
df.isnull().sum()
```

# Iniciamos la busqueda de datos atípicos (outlayers)






```python
#Ploteando Price para encontrar datos atípicos
df.boxplot(column="price",)
```

```python
#formula IQR para limpiar datos atípicos (outlayers)
q1 = df["price"].quantile(0.25)
q3 = df["price"].quantile(0.75)
iqr = q3 - q1

#Encontrando limites intercuartiles en price para sacar outliers
upper_limit = q3 + (1.5 * iqr)
lower_limit = q1 - (1.5 * iqr)
median = df["price"].median()

print("median", median)
print("upper limit", upper_limit)
print("lower limit", lower_limit)

df.loc[(df["price"] > upper_limit) | (df["price"] < lower_limit)]
```

```python
new_df = df.loc[(df['price'] < upper_limit) & (df['price'] > lower_limit)]
print('antes de remover outliers', len(df))
print('despues de remover outliers', len(new_df))
print('outliers removidos', len(df) - len(new_df))

new_df.boxplot(column="price", by=None, ax=None, grid=True)
```

```python
#Ploteando surface_covered para encontrar datos atípicos

plt.scatter(new_df['surface_covered'], new_df['rooms'], cmap='viridis')
plt.xlabel('Surface Covered')
plt.ylabel('Rooms')
plt.title('Scatter Plot of Surface Covered vs. Rooms')
plt.show()

```
